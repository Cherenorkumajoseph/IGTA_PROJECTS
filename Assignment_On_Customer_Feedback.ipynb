# CUSTOMER FEEDBACK ANALYSIS BETWEEN PYTHON AND EXCEL

# Importing the Necessary Libraries for Visualization

```python
import pandas as pd
import xlsxwriter as xls
print("Imported Libraries Successfully")
```

# Load Our Customer Feedback Dataset

```python
customer_data = pd.read_csv("customer_feedback_scores.csv")
customer_data.head(10)
```

# Checking Our Data if There are Non Values

```python
customer_data.isnull()
```

# Finding the Average Rating of Support Rating and Product Rating

```python
avg_rating = customer_data.groupby("Product") [["Support Rating", "Product Rating"]].mean().reset_index()
avg_rating.head()
```

# Create a New Excel Workbook and add a Bar Chart

```python
writer = pd.ExcelWriter("Feedback_Summary_with_Bar_Chart.xlsx", engine="xlsxwriter")
```

# Commiting the File to Excel

```python
avg_rating.to_excel(writer, sheet_name = "Feedback_Summary_with_Bar_Chart", index = False)
```

# Adding Bar Chart

```python
workbook = writer.book # Creates the Workbook Object
worksheet = writer.sheets["Feedback_Summary_with_Bar_Chart"] # Creates a Workbook Object

# Create a Bar Chart
chart = workbook.add_chart({"type": "column"})

chart.add_series({
    "name": "Support Rating",
    "categories": ["Feedback_Summary_with_Bar_Chart", 1, 0, len(avg_rating), 0],
    "values": ["Feedback_Summary_with_Bar_Chart", 1, 1, len(avg_rating), 2],
})

chart.add_series({
    "name": "Product Rating",
    "categories": ["Feedback_Summary_with_Bar_Chart", 1, 0, len(avg_rating), 0],
    "values": ["Feedback_Summary_with_Bar_Chart", 1, 2, len(avg_rating), 1],
})

chart.set_title({"name": "Average Cutomer Rating"})
chart.set_x_axis({"name": "Product"})
chart.set_y_axis({"name": "Average Rating"})

worksheet.insert_chart("F3", chart)
writer.close()
```
